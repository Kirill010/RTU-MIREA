```python
# Загрузить библиотеки

import numpy as np
from sklearn import preprocessing

# Создать признак
feature = np.array([[-500.5], [-100.1], [0], [100.1], [900.9]])
# Создать шкалировщик
minmax_scale = preprocessing.MinMaxScaler(feature_range=(0, 1))

# Прошкалировать признак
scaled_feature = minmax_scale.fit_transform(feature)

# Показать прошкалированный признак
scaled_feature
```

# 3.3.2. Задание

image.png

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

# Создать URL-адрес
url = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'
scaler1 = MinMaxScaler()
scaler2 = StandardScaler()
# Загрузить данные
dataframe = pd.read_csv(url)
dataframe[['sepal_length_cm']] = scaler1.fit_transform(dataframe[['sepal_length_cm']]) # минимакс преобразование
dataframe[['sepal_width_cm']] = scaler2.fit_transform(dataframe[['sepal_width_cm']]) # z - преобразование
dataframe
```
